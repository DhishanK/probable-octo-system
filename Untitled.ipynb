```python
def f(amount, interest, period):
    balance = amount
    T = 0
    pay = amount/period
    monthly = (interest/100) /12
    interest_list = []
    for x in range(1, period+1):
        i = amount * monthly
        interest_list.append(i)
        total = pay + i
        amount = amount - pay
        T += total
        
        
        
        #print(f"For Month: {x}, interest {i}, in need to pay: {total}. Remaining amount is:{amount}")
        #print(round(T,2))
    per_month = round(T/period, 2)
    print("Monthly: ", round(T/period,2))
    for i in range(period):
        monthly_interest = interest_list[i]
        monthly_principal = per_month-monthly_interest
        balance -= monthly_principal
        print("Principal: ", monthly_principal, "Interest: ", monthly_interest, "balance: ", balance)
```

```python
import plotly.graph_objects as go

def pmt_graph(amount, interest, months):
    monthly_rate = (interest / 100) / 12
    pmt = amount * (monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
    
    x_months = [0]
    remaining_balances = [round(amount, 2)]
    
    current_balance = amount
    for month in range(1, months + 1):
        interest_this_month = current_balance * monthly_rate
        principal_this_month = pmt - interest_this_month
        current_balance -= principal_this_month
        x_months.append(month)
        remaining_balances.append(round(current_balance, 2))

    biweekly_rate = (interest/100)/26
    biweekly_weeks = [0]
    biweekly_balances = [amount]
    current_balance = amount
    biweekly_period = 0

    while current_balance >0:
        biweekly_period += 1
        interest = current_balance * biweekly_rate
        principal  = pmt - interest
        
        if current_balance < principal:
            principal = current_balance

        current_balance -= principal
        m = biweekly_period * (12/26)
        biweekly_weeks.append(m)
        biweekly_balances.append(round(current_balance, 2))


    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=x_months,
        y=remaining_balances,
        mode='lines+markers',
    ))
    fig.add_trace(go.Scatter(
        x=biweekly_weeks,
        y=biweekly_balances,
        mode='lines+markers',
    ))
    fig.update_layout(
        title="Loan Amount over Time",
        xaxis_title="Timeline (Months)",
        yaxis_title="Amount ($)"
    )

    fig.show()
```
